```python
from sklearn.datasets import load_diabetes

# 당뇨병 데이터
d = load_diabetes()

print(d.data.shape, d.target.shape) # shape는 행렬의 모양(몇 곱하기 몇인지) 타겟값은 벡터값이라 행렬이 아니라서 뒷값이 없음
```

```python
x = d.data[:,2] # 두번째 독립변수만 가져옴
y = d.target
```

```python
# 산점도
import matplotlib.pyplot as plt
import numpy as np

plt.scatter(x, y) # 흩뿌리기(산점)
plt.xlabel("x"); plt.ylabel("y") # x, y축 이름 정하기
plt.axhline(y=np.mean(y), color="red") # y값 평균에 빨간줄 그리기
plt.show()
```
