```python
import urllib.request as req

# 이미지 다운로드
url = "https://images.unsplash.com/photo-1568967729548-e3dbad3d37e0?ixlib=rb-1.2.1&ixid=eyJhcHBfaWQiOjEyMDd9&auto=format&fit=crop&w=334&q=80"
req.urlretrieve(url, "test.png")

# OpenCV로 읽어들이기
import cv2
img = cv2.imread("test.png")
print(img)
```

```python
import matplotlib.pyplot as plt
import cv2

# 다운로드한 이미지 출력하기(그래프에 띄우기)
img = cv2.imread("test.png")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
